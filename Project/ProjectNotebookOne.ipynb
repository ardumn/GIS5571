## Develop Fluvial Network for Mars

```python
out_flow_direction_raster = arcpy.sa.FlowDirection("elevofmars", "NORMAL", None, "D8"); out_flow_direction_raster.save(r"C:\Users\Alexander Danielson\Desktop\Fall 2022Spring2023\ArcGIS I\FinalProject\FinalProject\Default.gdb\FlowDir_Mars") 
#Flow Direction of Mars
```

```python
out_accumulation_raster = arcpy.sa.FlowAccumulation("FlowDir_Mars", None, "FLOAT", "D8"); out_accumulation_raster.save(r"C:\Users\Alexander Danielson\Desktop\Fall 2022Spring2023\ArcGIS I\FinalProject\FinalProject\Default.gdb\FlowAcc_Mars") 
#Flow Accumulation of Mars with Float
```

```python
arcpy.ddd.Times("FlowAcc_Mars", "FlowDir_Mars", r"C:\Users\Alexander Danielson\Desktop\Fall 2022Spring2023\ArcGIS I\FinalProject\FinalProject\Default.gdb\Times_FlowAcc") 
#Times the Flow Direction and Flow Accumulation Raster together to produce part of the Flow Area
```

```python
out_raster = arcpy.ia.Log10("Times_FlowAcc"); out_raster.save(r"C:\Users\Alexander Danielson\Desktop\Fall 2022Spring2023\ArcGIS I\FinalProject\FinalProject\Default.gdb\Log10_TimesFlowDirecAcc")
# Creates 0 in cell values for calucalting river/stream order in following step for relcassifcation
```

```python
arcpy.ddd.Reclassify("Log10_TimesFlowDirecAcc", "VALUE", "0 1.850321 1;1.850321 6.740456 2", r"C:\Users\Alexander Danielson\Desktop\Fall 2022Spring2023\ArcGIS I\FinalProject\FinalProject\Default.gdb\Reclass_Log", "DATA")
# Relcassify for 4 classes for Stream order based on Strahler Order
```

```python
out_raster = arcpy.sa.StreamOrder("Reclass_Log", "FlowDir_Mars", "STRAHLER"); out_raster.save(r"C:\Users\Alexander Danielson\Desktop\Fall 2022Spring2023\ArcGIS I\FinalProject\FinalProject\Default.gdb\StreamOrder_Reclass")
```
